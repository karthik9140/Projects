# Task 3: Model Building

```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
updated_df = pd.read_csv('tel_churn.csv')
```

```python
updated_df.head()
```

```python
updated_df.drop(columns='Unnamed: 0',inplace=True)
```

```python
x = updated_df.drop(columns = 'Churn')
```

```python
x.head()
```

```python
y = updated_df['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
md = DecisionTreeClassifier()
md.fit(x_train,y_train)
```

```python
y_pred = md.predict(x_test)
```

```python
print(classification_report(y_test,y_pred,labels=[0,1]))
```

```python
from imblearn.combine import SMOTEENN
from sklearn.datasets import make_classification
```

```python
sm = SMOTEENN(random_state=42)
x_resampled, y_resampled = sm.fit_resample(x, y)
```

```python
xr_train,xr_test,yr_train,yr_test = train_test_split(x_resampled,y_resampled,test_size = 0.2)

md2 = DecisionTreeClassifier()
md2.fit(xr_train,yr_train)
```

```python
yr_pred = md2.predict(xr_test)
```

```python
print(classification_report(yr_test,yr_pred,labels =[0,1]))
```
